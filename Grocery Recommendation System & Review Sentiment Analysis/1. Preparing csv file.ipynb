<h1><center> Grocery and Gourment Food Amazon </center></h1>

Data yang digunakan dalam projek ini adalah data review dan rating dari produk grocery dan gourment food yang dijual di Amazon.

Data diperoleh dari website http://jmcauley.ucsd.edu/data/amazon/links.html. Terdapat dua data utama yang digunakan yaitu rating dan metadata dari produk.

<img src='grocery.jpg' style='height:400px'>

Terdapat lebih dari 8500 produk yang direview dalam data ini dan terdapat lebih dari 150000 review. Data ini cocok digunakan untuk melakukan sentiment analysis terhadap dari review yang diberikan. Selain itu, data ini juga cocok untuk digunakan sebagai data dalam pembuatanan sistem rekomendasi.

```python
# import package atau modul standar yang dibutuhkan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import package dan modul untuk membaca file json
from pandas.io.json import json_normalize
from ast import literal_eval
```

**Membaca file review bertipe json dan menyimpannya ke file tipe csv**

Data yang didapatkan masih berupa tipe json, kita perlu mengubahnya terlebih dahulu menjadi tipe file csv. file yang bertipe csv lebih mudah untuk dibaca menggunakan method dari pandas yaitu read_csv

```python
data = []
with open('Grocery_and_Gourmet_Food_5.json') as f:
    for i, line in enumerate(f):
        data.append(literal_eval(line))
        
data[:3]
```

```python
review = json_normalize(data)

review.head()
```

```python
# menyimpan file review dalam bentuk csv
review.to_csv('review.csv', index=False)
```

Data review sudah disimpan ke dalam bentuk csv. Kita akan menggunakan file ini untuk menganalisis lebih jauh dari data review.

**Membaca file meta bertipe json dan menyimpannya ke file tipe csv**

Meta merupakan file yang berisi tentang informasi produk. Sama halnya dengan data review, meta juga masih dalam betuk json dan kita perlu mengubahnya ke dalam bentuk csv.

```python
meta_data = []
with open('meta_Grocery_and_Gourmet_Food.json') as f:
    for i, line in enumerate(f):
        meta_data.append(literal_eval(line))

meta_data[:3]
```

```python
pd.options.display.max_columns = 200
```

```python
meta = json_normalize(meta_data)

meta.head()
```

```python
#menyimpan file meta dalam bentuk csv
meta.to_csv('meta.csv', index=False)
```

ast.literal_eval digunakan untuk mengubah string menjadi suatu objek di python. Misalkan, ketika menggunakan literal eval terhadap '[1,2]' maka akan menjadi list [1,2]

Sementara itu, json_normalize berfungsi untuk menormalisasi json data menjadi flat-table.
